```python
import json
import pickle
# Helper function to load token list from pickle file and load topic words from json file
def load_token_list_and_topics(token_file='token_list.pkl', topics_file='topic_words.json'):
    # Load token list from pickle
    with open('PickleFiles/Tokens/' + token_file, 'rb') as f:
        token_list = pickle.load(f)

    # Load topic words from JSON
    with open('PickleFiles/Tokens/' + topics_file, 'r') as f:
        topic_words = json.load(f)

    return token_list, topic_words
```

```python
###Quick note about the order of the models: the order of the models goes in alphabetical order of the file names, so all_corpus_again would be before fan_corpus_only. For every model, all_configs is first, followed by stop_words, followed by hdbscan and finally finished with umap
```

```python
token_list, topic_words = load_token_list_and_topics('all_corpus_again_tokens_all_configs.pkl',
                                                     'all_corpus_again_topics_all_configs.json')

from octis.evaluation_metrics.coherence_metrics import Coherence

#Initialize coherence metric model
coherence_model = Coherence(texts=token_list,
                            topk=10,
                            measure='c_v')

coherence_score = coherence_model.score({"topics": topic_words})
print("Coherence (C_v:)", coherence_score)
```
